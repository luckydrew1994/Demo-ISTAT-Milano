```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carica il dataset ISTAT (adatta il nome del file se diverso)
df = pd.read_csv("DCIS_POPRES.csv", sep=";", skiprows=2)

# 2. Rinomina le colonne
df.columns = ["Codice", "Comune", "Eta", "Maschi", "Femmine", "Totale"]

# 3. Filtra un comune specifico (es. Abbiategrasso)
df_comune = df[df["Comune"] == "Abbiategrasso"].copy()

# 4. Mostra la tabella con stile elegante
df_comune_styled = df_comune.head(20).style.set_caption("ðŸ“Š Dati demografici - Abbiategrasso (prime 20 etÃ )") \
    .background_gradient(subset=["Maschi", "Femmine", "Totale"], cmap="Blues") \
    .format({"Maschi":"{:.0f}", "Femmine":"{:.0f}", "Totale":"{:.0f}"})

df_comune_styled
```

```python
# 5. Calcola percentuali rispetto al totale del comune
totale = df_comune["Totale"].sum()
df_comune["PercMaschi"] = df_comune["Maschi"] / totale * 100
df_comune["PercFemmine"] = df_comune["Femmine"] / totale * 100

# 6. Inverti i maschi per la piramide (sinistra)
df_comune["PercMaschi"] = -df_comune["PercMaschi"]

# 7. Grafico piramide delle etÃ 
plt.figure(figsize=(10,8))
plt.barh(df_comune["Eta"], df_comune["PercMaschi"], color="#1f77b4", label="Maschi")
plt.barh(df_comune["Eta"], df_comune["PercFemmine"], color="#ff7f0e", label="Femmine")

plt.title("Piramide delle etÃ  - Abbiategrasso (2025)")
plt.xlabel("Percentuale della popolazione (%)")
plt.ylabel("EtÃ ")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Carica il dataset ISTAT
df = pd.read_csv("DCIS_POPRES.csv", sep=";", skiprows=2)
df.columns = ["Codice", "Comune", "Eta", "Maschi", "Femmine", "Totale"]

# Filtra Abbiategrasso
df_comune = df[df["Comune"] == "Abbiategrasso"].copy()
```

```python
# Salva in CSV
df_comune.to_csv("docs/Abbiategrasso_tabella.csv", index=False)

# Salva in Excel
df_comune.to_excel("docs/Abbiategrasso_tabella.xlsx", index=False)

# Salva in HTML con stile
html = df_comune.head(20).style.set_caption("ðŸ“Š Abbiategrasso (prime 20 etÃ )") \
    .background_gradient(subset=["Maschi","Femmine","Totale"], cmap="Blues") \
    .format({"Maschi":"{:.0f}","Femmine":"{:.0f}","Totale":"{:.0f}"}).to_html()

with open("docs/Abbiategrasso_tabella.html", "w", encoding="utf-8") as f:
    f.write(html)
```
